Part II    
Obtain three numbers that sum to 2020

```python
target = 2020
sorted_nums = sorted(numbers)
for i, num in enumerate(sorted_nums):
    l = i + 1
    j = len(sorted_nums) - 1
    while l < j:
        curr_sum = sorted_nums[i] + sorted_nums[l] + sorted_nums[j]
        if curr_sum > target:
            j -= 1
        elif curr_sum < target:
            l += 1
        else:
            print(sorted_nums[i] * sorted_nums[l] * sorted_nums[j])
            break
```
